#4. Jumlah Customer

a. Berdasarkan gender

```python
#Menunjukkan jumlah customer berdasarkan gender

df1=df[['customerID','gender']].groupby(by='gender').count()
```

```python
#Mengurutkan hasil dari jumlah terbesar

df1.sort_values('customerID',ascending=True)
```

b. Berdasarkan InternetService

```python
#Menunjukkan jumlah customer berdasarkan InternetService

df2=df[['customerID','InternetService']].groupby(by='InternetService').count()
```

```python
#Mengurutkan hasil dari jumlah terbesar

df2.sort_values('customerID',ascending=True)
```

c. Berdasarkan PaymentMethod

```python
#Menunjukkan jumlah customer berdasarkan PaymentMethod

df3=df[['customerID','PaymentMethod']].groupby(by='PaymentMethod').count()
```

```python
#Mngurutkan hasil dari jumlah terbesar

df3.sort_values('customerID',ascending=True)
```

#5. Membuat kolom baru golongan *MonthlyCharges*

```python
#Membuat kolom baru ChargesType
df['ChargesType']=pd.cut(
    df['MonthlyCharges'],
    bins=[0,35.5,70.5,89.9,999],
    labels=['Low','Medium','High','Very High'])
```
